```python
# Import libraries
from transformers import pipeline
```

```python
# Load a pre-trained conversational model (e.g., DialoGPT)
chatbot = pipeline("text-generation", model="microsoft/DialoGPT-medium")
```

```python
# Define a function to handle customer queries
def customer_support_chatbot(query):
    # Generate a response using the chatbot model
    response = chatbot(query, max_length=50, num_return_sequences=1)[0]['generated_text']
    return response
```

```python
# Example usage
queries = [
    "How can I book a ticket?",
    "My train is delayed. What should I do?",
    "How do I request a refund?",
    "What are the ticket prices?"
]

for query in queries:
    print(f"Customer: {query}")
    response = customer_support_chatbot(query)
    print(f"Chatbot: {response}\n")
```

```python
# Define a function to check refund eligibility
def is_refund_eligible(ticket_type, delay_duration, purchase_type):
    """
    Determine if a refund is eligible based on rules.
    :param ticket_type: Type of ticket (e.g., 'Standard', 'First Class')
    :param delay_duration: Delay duration in minutes
    :param purchase_type: Purchase type (e.g., 'Online', 'Station')
    :return: Boolean indicating refund eligibility
    """
    # Refund rules
    if ticket_type == "First Class" and delay_duration > 30:
        return True
    elif ticket_type == "Standard" and delay_duration > 60:
        return True
    elif purchase_type == "Online" and delay_duration > 45:
        return True
    else:
        return False

# Define a function to process refunds
def process_refund(ticket_id, ticket_type, delay_duration, purchase_type):
    """
    Process a refund if eligible.
    :param ticket_id: Unique ticket ID
    :param ticket_type: Type of ticket
    :param delay_duration: Delay duration in minutes
    :param purchase_type: Purchase type
    :return: Refund status message
    """
    if is_refund_eligible(ticket_type, delay_duration, purchase_type):
        return f"Refund processed for Ticket ID: {ticket_id}. Amount will be credited within 5-7 business days."
    else:
        return f"Refund not eligible for Ticket ID: {ticket_id}. Delay duration or ticket type does not meet refund criteria."

# Example usage
tickets = [
    {"ticket_id": 101, "ticket_type": "First Class", "delay_duration": 45, "purchase_type": "Online"},
    {"ticket_id": 102, "ticket_type": "Standard", "delay_duration": 70, "purchase_type": "Station"},
    {"ticket_id": 103, "ticket_type": "Standard", "delay_duration": 30, "purchase_type": "Online"}
]

for ticket in tickets:
    result = process_refund(
        ticket["ticket_id"],
        ticket["ticket_type"],
        ticket["delay_duration"],
        ticket["purchase_type"]
    )
    print(result)
```

```python
# Define a function to handle customer queries
def handle_customer_query(query, ticket_id=None, ticket_type=None, delay_duration=None, purchase_type=None):
    if "refund" in query.lower() and ticket_id and ticket_type and delay_duration and purchase_type:
        # Process refund request
        refund_status = process_refund(ticket_id, ticket_type, delay_duration, purchase_type)
        return refund_status
    else:
        # Handle general queries using the chatbot
        return customer_support_chatbot(query)

# Example usage
query = "My train was delayed by 50 minutes. Can I get a refund?"
ticket_id = 101
ticket_type = "First Class"
delay_duration = 50
purchase_type = "Online"

response = handle_customer_query(query, ticket_id, ticket_type, delay_duration, purchase_type)
print(response)
```
